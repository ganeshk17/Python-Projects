```python
GUESS GAME

import random
randomNo=random.randint(1,100)
Userguess=0
guess=0
while Userguess!=randomNo:
    Userguess=int(input('enter the no.'))
    guess+=1
    if Userguess==randomNo:
        print(f'you won in {guess} guesses')
        
    else:
        if Userguess>randomNo:
            print('enter smaller no.')
        else:
            print('enter larger no.')
```
