```python


class train:
    def __init__(self,name,fare,seats):
        self.name=name
        self.fare=fare
        self.seats=seats

    def trainstatus(self):
        print(f'name of train is,{self.name}')
        print(f'fare is{self.fare}')
        print(f'seats available are{self.seats}')
    
    def bookticket(self):
        if self.seats>0:
            print(f'your ticket is booked{self.seats}')
            self.seats=self.seats-1
        else:
            print('no seats available')
    

railway=train('rajdhani',1000,20)
railway.trainstatus()
railway.bookticket()
railway.trainstatus()
```
